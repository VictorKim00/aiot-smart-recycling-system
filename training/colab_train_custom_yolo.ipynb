# Custom YOLO Training for AIoT Smart Recycling
Train on laptop/Colab, then copy `best.pt` to Raspberry Pi.


````python
# Colab only. Skip if your laptop already has ultralytics.
!pip install -q ultralytics
````

Upload or mount your dataset so the structure is:
```
datasets/waste/images/train
datasets/waste/images/val
datasets/waste/labels/train
datasets/waste/labels/val
datasets/waste/data.yaml
```


````python
from ultralytics import YOLO
model = YOLO('yolov8n.pt')
model.train(data='datasets/waste/data.yaml', epochs=80, imgsz=640, batch=16, device=0, project='runs/waste', name='waste_custom', plots=True)
````

````python
model = YOLO('runs/waste/waste_custom/weights/best.pt')
metrics = model.val(data='datasets/waste/data.yaml', imgsz=640, device=0)
print(metrics)
````

````python
model.predict(source='datasets/waste/images/val', imgsz=640, conf=0.25, save=True, save_txt=True)
````

Download `runs/waste/waste_custom/weights/best.pt` and copy it to Raspberry Pi project folder: `models/best.pt`.
